```python
import openmc.deplete as dep
import openmc.data as data
import os
```

```python
data.endf.convert_endf(
    source_dir=os.path.expanduser("~/endfb-vii.1/"),
    destination=os.path.expanduser("~/openmc_data/endfb71_hdf5"),
    library="endfb71",
    particles=["neutron", "decay", "neutron_fission_yield"]
)
```

```python
dep.make_chain(
    decay="~/openmc_data/endfb71_hdf5/decay.xml",
    fpy="~/openmc_data/endfb71_hdf5/fission_yields.xml",
    nfy="~/openmc_data/endfb71_hdf5/reactions.xml",
    chain_file="~/openmc_data/chain_endfb71.xml"
)
```

```python
import openmc.deplete

# Paths to your ENDF sublibrary XMLs
decay_files = ["~/nndc_hdf5/decay.xml"]
fpy_files   = ["~/openmc_data/endf-b-vii.1/fission_yields.xml"]
neutron_files = ["~/openmc_data/endfb71_hdf5/neutrons.xml"]

# Build the chain
chain = openmc.deplete.Chain.from_endf(
    decay_files=decay_files,
    fpy_files=fpy_files,
    neutron_files=neutron_files,
)

# Export to XML for depletion use
chain.export_to_xml("~/openmc_data/chain_endfb71.xml")
print("Depletion chain written to chain_endfb71.xml")
```
